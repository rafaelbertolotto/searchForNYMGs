```python
common_path = 'builded_or_modified_cat/'

# This is GOG final field with RVs
gogRVFinalField = pd.read_csv(common_path+'gogRVFinalField.csv')

# This is GOG final field for propper motion analysis 
gogFinalField   = pd.read_csv(common_path+'gogFinalField.csv')

# This all of GOG cleaned
gogClean        = pd.read_csv(common_path+'gog_XYZUVW.csv')

# This is all of GDR3 cleaned
gdr3Clean       = pd.read_csv(common_path+'gdr3_XYZUVW.csv')

# This is all of GDR# MS and PMS cleaned
gdr3MSPMS       = pd.read_csv(common_path+'gdr3MSPMS.csv')

# This is all of GDR3 MS cleaned
gdr3MS          = pd.read_csv(common_path+'gdr3MS.csv')

# This is all of GDR3 MS and PMS with RVs cleaned
gdr3RVMSPMS     = pd.read_csv(common_path+'gdr3RVMSPMS.csv')

# This is all of GDR3 MS with RVs cleaned
gdr3RVMS        = pd.read_csv(common_path+'gdr3RVMS.csv')

```

```python
def updateForLSR(data,filename):
    
    dataToChange = data[(data.U.isnull().values == False) & (data.pmra.isnull().values == True)]
    
    if dataToChange.shape[0] > 0:
        Galactic_coord               = coordinates.Galactic(u                   = dataToChange.X.to_numpy()*u.pc,
                                                            v                   = dataToChange.Y.to_numpy()*u.pc, 
                                                            w                   = dataToChange.Z.to_numpy()*u.pc,
                                                            U                   = dataToChange.U.to_numpy()*u.km/u.s,
                                                            V                   = dataToChange.V.to_numpy()*u.km/u.s,
                                                            W                   = dataToChange.W.to_numpy()*u.km/u.s,
                                                            representation_type = CartesianRepresentation,
                                                            differential_type   = CartesianDifferential)  
            
        icrs                         = Galactic_coord.transform_to(ICRS)
        dataToChange.parallax        = pd.DataFrame(1000/icrs.distance,index=dataToChange.index)
        dataToChange.ra              = pd.DataFrame(icrs.ra,index=dataToChange.index)
        dataToChange.dec             = pd.DataFrame(icrs.dec,index=dataToChange.index)
        dataToChange.radial_velocity = pd.DataFrame(icrs.radial_velocity,index=dataToChange.index)
        dataToChange.pmra            = pd.DataFrame(icrs.pm_ra_cosdec,index=dataToChange.index)
        dataToChange.pmdec           = pd.DataFrame(icrs.pm_dec,index=dataToChange.index)
    
        data = pd.concat([data[(data.U.isnull().values == True) | (data.pmra.isnull().values == False)],dataToChange],axis=0)
    
    dataToChange = data[data.ra.isnull().values == True]
    
    if dataToChange.shape[0] > 0:
        Galactic_coord               = coordinates.Galactic(u                   = dataToChange.X.to_numpy()*u.pc,
                                                            v                   = dataToChange.Y.to_numpy()*u.pc, 
                                                            w                   = dataToChange.Z.to_numpy()*u.pc,
                                                            representation_type = CartesianRepresentation,
                                                            differential_type   = CartesianDifferential)  
            
        icrs                         = Galactic_coord.transform_to(ICRS)
        dataToChange.parallax        = pd.DataFrame(1000/icrs.distance,index=dataToChange.index)
        dataToChange.ra              = pd.DataFrame(icrs.ra,index=dataToChange.index)
        dataToChange.dec             = pd.DataFrame(icrs.dec,index=dataToChange.index)
    
        data = pd.concat([data[data.ra.isnull().values == False],dataToChange],axis=0)
        
    data = data.reset_index(drop=True)
    # We compute the LSR proper motions, to do so, we set all RVs NaNs to 0
    icrs            = coordinates.ICRS(ra              = data["ra"].to_numpy()*u.degree,
                                       dec             = data["dec"].to_numpy()*u.degree,
                                       distance        = (1000/data["parallax"]).to_numpy()*u.pc,
                                       pm_ra_cosdec    = data["pmra"].to_numpy()*u.mas/u.yr,
                                       pm_dec          = data["pmdec"].to_numpy()*u.mas/u.yr,
                                       radial_velocity = data["radial_velocity"].fillna(0).to_numpy()*u.km/u.s)

    # We transform from ICRS to LSR
    lsr                                = icrs.transform_to(LSR())
    data['pmraLSR']                    = pd.DataFrame(lsr.pm_ra_cosdec)
    data['pmdecLSR']                   = pd.DataFrame(lsr.pm_dec)
    
    data.to_csv(common_path+filename,index=False)
```

```python
updateForLSR(gogRVFinalField,'gogRVFinalField.csv')
updateForLSR(gogFinalField,'gogFinalField.csv')
updateForLSR(gogClean,'gog_XYZUVW.csv')
updateForLSR(gdr3Clean,'gdr3_XYZUVW.csv')
updateForLSR(gdr3MSPMS,'gdr3MSPMS.csv')
updateForLSR(gdr3MS,'gdr3MS.csv')
updateForLSR(gdr3RVMSPMS,'gdr3RVMSPMS.csv')
updateForLSR(gdr3RVMS,'gdr3RVMS.csv')
```
